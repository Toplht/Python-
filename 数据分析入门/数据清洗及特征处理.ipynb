#### 1.2对缺省值进行处理
（1）处理缺省值一般有几种思路

```python
df[df['Age'] == None] = 0
df.head(3)
```

```python
df[df['Age'].isnull()] = 0
df.head(3)
```

```python
df[df['Age'] == np.nan] = 0
df.head()
```

> 数值列读取数据后，空缺值的数据类型为float64所以用None一般索引不到，比较的时候最好用np.nan
